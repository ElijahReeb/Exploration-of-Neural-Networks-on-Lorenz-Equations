```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rcParams
from scipy import integrate
from mpl_toolkits.mplot3d import Axes3D
```

```python
def lorenz_solve(rho):

#     rcParams.update({'font.size': 18})
#     plt.rcParams['figure.figsize'] = [12, 12]

    dt = 0.01
    T = 8
    t = np.arange(0,T+dt,dt)
    beta = 8/3
    sigma = 10
    rho = rho


    lor_input = np.zeros((100*(len(t)-1),3))
    lor_output = np.zeros_like(lor_input)

#     fig,ax = plt.subplots(1,1,subplot_kw={'projection': '3d'})
#     title = 'Lorenz Equation Output for rho =  {} '.format(rho)
#     fig.suptitle(title, fontsize = 20)


    def lorenz_deriv(x_y_z, t0, sigma=sigma, beta=beta, rho=rho):
        x, y, z = x_y_z
        return [sigma * (y - x), x * (rho - z) - y, x * y - beta * z]

    np.random.seed(123)
    x0 = -15 + 30 * np.random.random((100, 3))

    x_t = np.asarray([integrate.odeint(lorenz_deriv, x0_j, t)
                      for x0_j in x0])

    for j in range(100):
        lor_input[j*(len(t)-1):(j+1)*(len(t)-1),:] = x_t[j,:-1,:]
        lor_output[j*(len(t)-1):(j+1)*(len(t)-1),:] = x_t[j,1:,:]
        x, y, z = x_t[j,:,:].T
    
#         ax.plot(x, y, z,linewidth=1)
#         ax.scatter(x0[j,0],x0[j,1],x0[j,2],color='r')
#     ax.view_init(18, -113)
#     plt.show()

    return lor_input, lor_output
 
```

### Part 1
Train a NN to advance the solution from t to t + ∆t for ρ = 10, 28 and 40. Now see how well
your NN works for future state prediction for ρ = 17 and ρ = 35

```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
```

```python
# Define activation functions
def logsig(x):
    return 1 / (1 + torch.exp(-x))

def radbas(x):
    return torch.exp(-torch.pow(x, 2))

def purelin(x):
    return x

# Define the model
class MyModel(nn.Module):
    def __init__(self):
        super(MyModel, self).__init__()
        self.fc1 = nn.Linear(in_features=3, out_features=40)
        self.fc2 = nn.Linear(in_features=40, out_features=30)
        self.fc3 = nn.Linear(in_features=30, out_features=10)
        self.fc4 = nn.Linear(in_features=10, out_features=3)
        
    def forward(self, x):
        x = logsig(self.fc1(x))
        x = radbas(self.fc2(x))
        x = radbas(self.fc3(x))
        x = purelin(self.fc4(x))
        return x

# Create model instance
model = MyModel()

# Define loss function and optimizer
criterion = nn.CrossEntropyLoss()
optimizer = optim.SGD(model.parameters(), lr=0.02, momentum=.4)
```

```python
# Create training data for rho = 10, 28, and 40
# this is done by running the function above to generate the lorenz data

rho10_inp, rho10_out = lorenz_solve(10)
rho28_inp, rho28_out = lorenz_solve(28)
rho40_inp, rho40_out = lorenz_solve(40)
print(np.shape(rho10_inp))
print(np.shape(rho10_out))

# compile 3 datasets into 1 dataset for training
train_data = np.concatenate((rho10_inp, rho28_inp, rho40_inp), axis=0)
print(np.shape(train_data))

train_labels = np.concatenate((rho10_out, rho28_out, rho40_out), axis=0)
print(np.shape(train_labels))
# Convert numpy arrays to PyTorch tensors

#nn_input = torch.from_numpy(nn_input).float()
#nn_output = torch.from_numpy(nn_output).float()

nn_input = torch.tensor(train_data, dtype=torch.float32)
nn_output = torch.tensor(train_labels, dtype =torch.float32)
```

```python
# Train the model
num_epochs = 20

loss_graph = np.zeros(int(num_epochs))

for epoch in range(num_epochs):
    optimizer.zero_grad()
    outputs = model(nn_input)
    loss = criterion(outputs, nn_output)
    loss.backward()
    optimizer.step()
    print(f"Epoch {epoch+1}, loss={loss.item():.4f}")
    loss_graph[epoch] = loss.item()

eps = np.arange(num_epochs)
plt.plot(eps, loss_graph, 'r', label = 'Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('FFNN Model on Lorenz Equations Loss')
plt.legend()
plt.show()
```

```python
# Test the network on rho of 17 and 35

rho17_inp, rho17_out = lorenz_solve(17)
rho35_inp, rho35_out = lorenz_solve(35)

test17_inp = torch.tensor(rho17_inp, dtype=torch.float32)
test17_out = torch.tensor(rho17_out, dtype =torch.float32)

# Evaluate the model on the test data
with torch.no_grad():
    y_pred = model(test17_inp)
    loss = nn.MSELoss()(y_pred, test17_out)

# Compute the accuracy
acc = 1 - loss.item()

# Print the accuracy
print("Test accuracy on rho = 17:", acc)

# Evaluate on rho = 35

test35_inp = torch.tensor(rho35_inp, dtype=torch.float32)
test35_out = torch.tensor(rho35_out, dtype =torch.float32)

# Evaluate the model on the test data
with torch.no_grad():
    y_pred = model(test35_inp)
    loss = nn.MSELoss()(y_pred, test35_out)

# Compute the accuracy
acc = 1 - loss.item()

# Print the accuracy
print("Test accuracy on rho = 35:", acc)

```

### Part 2
Compare feed-forward, LSTM, RNN and Echo State Networks for forecasting the dynamics.

```python
# Now to set up LSTM Model for this data

import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset

# Define the LSTM model
class LSTM(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers, output_size):
        super(LSTM, self).__init__()
        self.hidden_size = hidden_size
        self.num_layers = num_layers
        self.lstm = nn.LSTM(input_size, hidden_size, num_layers, batch_first=True)
        self.fc = nn.Linear(hidden_size, output_size)

    def forward(self, x):
        h0 = torch.zeros(self.num_layers, x.size(0), self.hidden_size).to(x.device)
        c0 = torch.zeros(self.num_layers, x.size(0), self.hidden_size).to(x.device)
        out, _ = self.lstm(x, (h0, c0))
        out = self.fc(out[:, -1, :])
        return out

# Define hyperparameters
input_size = 3
hidden_size = 16
num_layers = 2
output_size = 3
lr = 0.01
num_epochs = 100

# Define the LSTM model and optimizer
lstmmodel = LSTM(input_size, hidden_size, num_layers, output_size)
optimizer = optim.Adam(model.parameters(), lr=lr)
criterion = nn.MSELoss()
```

```python
# Create a PyTorch DataLoader object for the training and test data from rho of 10,28,40
train_data = TensorDataset(nn_input, nn_output)
train_loader = DataLoader(train_data, batch_size=16, shuffle=True)

# Create 2 different loaders for rho = 17 and 35

test17_data = TensorDataset(test17_inp, test17_out)
test17_loader = DataLoader(test17_data, batch_size=16, shuffle=False)

test35_data = TensorDataset(test35_inp, test35_out)
test35_loader = DataLoader(test35_data, batch_size=16, shuffle=False)

# Train the model
for epoch in range(num_epochs):
    lstmmodel.train()
    train_loss = 0
    for i, (inputs, labels) in enumerate(train_loader):
        optimizer.zero_grad()
        outputs = lstmmodel(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        train_loss += loss.item()

    # Evaluate the model on the test set
    lstmmodel.eval()
    with torch.no_grad():
        test_loss = 0
        for inputs, labels in test17_loader:
            outputs = lstmmodel(inputs)
            loss = criterion(outputs, labels)
            test_loss += loss.item()

    # Print the training and test loss for each epoch
    print(f"Epoch {epoch+1} - Train Loss: {train_loss/i:.4f}, Test Loss: {test_loss/len(test_loader):.4f}")

# Evaluate the model on the rho = 17 test set
lstmmodel.eval()
with torch.no_grad():
    test_loss = 0
    for inputs, labels in test17_loader:
        outputs = lstmmodel(inputs)
        loss = criterion(outputs, labels)
        test_loss += loss.item()

    # Print the test loss and accuracy
    print(f"Test Loss: {test_loss/len(test17_loader):.4f}")
    
# Evaluate the model on the rho = 35 test set
lstmmodel.eval()
with torch.no_grad():
    test_loss = 0
    for inputs, labels in test35_loader:
        outputs = lstmmodel(inputs)
        loss = criterion(outputs, labels)
        test_loss += loss.item()

    # Print the test loss and accuracy
    print(f"Test Loss: {test_loss/len(test35_loader):.4f}")
```

```python
# Now to set up RNN Model for this data

# Define the RNN model
class RNN(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super(RNN, self).__init__()
        self.hidden_size = hidden_size
        self.i2h = nn.Linear(input_size + hidden_size, hidden_size)
        self.i2o = nn.Linear(input_size + hidden_size, output_size)
        self.sigmoid = nn.Sigmoid()

    def forward(self, input_tensor, hidden_tensor):
        combined = torch.cat((input_tensor, hidden_tensor), 1)
        hidden = self.i2h(combined)
        output = self.i2o(combined)
        output = self.sigmoid(output)
        return output, hidden

    def init_hidden(self):
        return torch.zeros(1, self.hidden_size)

# Define the training function
def train_rnn(model, inputs, targets, criterion, optimizer, epochs):
    for epoch in range(epochs):
        loss = 0
        hidden = model.init_hidden()

        for i in range(len(inputs)):
            input_tensor = torch.tensor(inputs[i], dtype=torch.float32).view(1, -1)
            target_tensor = torch.tensor(targets[i], dtype=torch.float32).view(1, -1)

            optimizer.zero_grad()

            output, hidden = model(input_tensor, hidden)
            loss += criterion(output, target_tensor)

            loss.backward(retain_graph=True)
            optimizer.step()

        print('Epoch [{}/{}], Loss: {:.4f}'.format(epoch+1, epochs, loss.item()))

# Define the testing function
def test_rnn(model, inputs):
    hidden = model.init_hidden()

    for i in range(len(inputs)):
        input_tensor = torch.tensor(inputs[i], dtype=torch.float32).view(1, -1)

        output, hidden = model(input_tensor, hidden)

    return output.detach().numpy()
```

```python
# Define the training and testing data
inputs = nn_input
targets = nn_output

# Define the model, criterion, and optimizer
input_size = 3
hidden_size = 2
output_size = 3

RNNmodel = RNN(input_size, hidden_size, output_size)
criterion = nn.MSELoss()
optimizer = optim.Adam(RNNmodel.parameters(), lr=0.01)

# Train the model
epochs = 100
train_rnn(RNNmodel, inputs, targets, criterion, optimizer, epochs)

# Test the model
test17_inputs = test17_inp
test35_inputs = test35_inp

output17 = test_rnn(RNNmodel, test17_inputs)
print("Predicted output:", output17)


output35 = test_rnn(RNNmodel, test35_inputs)
print("Predicted output:", output35)
```

```python
# Now to set up Echo State Network for this data

class ESN(nn.Module):
    def __init__(self, input_size, hidden_size, output_size, sparsity=0.1, spectral_radius=0.9):
        super(ESN, self).__init__()
        
        # Define ESN parameters
        self.input_size = input_size
        self.hidden_size = hidden_size
        self.output_size = output_size
        self.sparsity = sparsity
        self.spectral_radius = spectral_radius
        
        # Initialize ESN weights
        self.input_weights = torch.randn((hidden_size, input_size))
        self.hidden_weights = torch.randn((hidden_size, hidden_size))
        self.output_weights = torch.randn((output_size, hidden_size))
        
        # Make ESN weights sparse and ensure spectral radius is within desired range
        mask = torch.rand_like(self.input_weights)
        mask = mask < sparsity
        self.input_weights *= mask.float()
        self.hidden_weights *= spectral_radius / torch.max(torch.abs(torch.linalg.eig(self.hidden_weights)[0]))
        
        # Initialize ESN biases
        self.hidden_bias = torch.randn((hidden_size,))
        self.output_bias = torch.randn((output_size,))
        
        # Define ESN activation function
        self.activation = nn.Tanh()
        
    def forward(self, x):
        # Initialize ESN hidden state
        h = torch.zeros((self.hidden_size,))
        
        # Iterate over input sequence
        for i in range(x.shape[0]):
            # Compute ESN output and update hidden state
            u = torch.matmul(self.input_weights, x[i]) + torch.matmul(self.hidden_weights, h) + self.hidden_bias
            h = self.activation(u)
            y = torch.matmul(self.output_weights, h) + self.output_bias
            
        return y
    
def train_esn(model, x, y, optimizer, criterion):
    # Set model to train mode
    model.train()
    
    # Reset optimizer gradients
    optimizer.zero_grad()
    
    # Run ESN on input sequence
    y_pred = model(x)
    
    # Compute loss and gradients
    loss = criterion(y_pred, y)
    loss.backward()
    
    # Update weights
    optimizer.step()
    
    return loss.item()

def test_esn(model, x, y):
    # Set model to evaluation mode
    model.eval()
    
    # Run ESN on input sequence
    y_pred = model(x)
    
    # Compute accuracy
    accuracy = torch.mean(torch.abs(y_pred - y))
    
    return accuracy.item()
```

```python
input_size = 3
hidden_size = 2
output_size = 3

ESNmodel = ESN(input_size, hidden_size, output_size)

criterion = nn.MSELoss()
optimizer = optim.Adam(ESNmodel.parameters(), lr=0.01)

# Train the model
epochs = 100
train_esn(ESNmodel, inputs, targets, criterion, optimizer, epochs)

# Test the model
test17_inputs = test17_inp
test35_inputs = test35_inp

output17 = test_esn(ESNmodel, test17_inputs)
print("Predicted output:", output17)


output35 = test_esn(ESNmodel, test35_inputs)
print("Predicted output:", output35)
```
